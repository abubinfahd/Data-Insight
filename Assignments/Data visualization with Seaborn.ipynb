# How to start data visualization with Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

## Why we use Seaborn?<br>
1.Easy to use<br>
2.Works well with Pandas data structure<br>
3.Built on Matplotlib<br>

## Getting started

We should import matplotlib along with seaborn.Because of seaborn build on matplotlib.

```python
import seaborn as sns #Samuel Norman Seaborn ( sns )
import matplotlib.pyplot as plt
```

## Scatter Plot

```python
# random data
height = [56, 67, 70, 71, 75, 57, 67]
weight = [120, 136, 123, 140, 172, 155,200]
sns.scatterplot(x=height, y=weight)
plt.show()
```

# Count plot

```python
gender = ["Female", "Female", "Female", "Male", "Male", "Male", "Male"]
sns.countplot(x=gender)
plt.show()
```

Seaborn provides an API on top of Matplotlib that offers sane choices for plot style and color defaults, defines simple high-level functions for common statistical plot types, and integrates with the functionality provided by Pandas DataFrame. 

```python
import pandas as pd
mpg = pd.read_csv("mpg.csv")
mpg.head()
```

# Scatter plot

```python
sns.scatterplot(x="mpg",y="horsepower",data=mpg)
plt.show()
```

## Scatterplot with hue

```python
# In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding
sns.scatterplot(x="mpg",y="horsepower",data=mpg, hue="horsepower")
plt.show()
```

```python
# In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding
sns.scatterplot(x="mpg",y="horsepower",data=mpg, hue="mpg")
plt.show()
```

Look at these two graphs carefully with legends

## Specifying hue colors

```python
# In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding
hue_colors = {
    "usa":"black",
    "europe":"red",
    "japan":"blue",
}
sns.countplot(x="origin",data=mpg, hue="origin", palette=hue_colors)
plt.show()
```
